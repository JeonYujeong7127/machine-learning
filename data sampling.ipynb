```python
import json
```

add(file_path) 함수로 먼저 10개의 분야에서 10개의 파일을 읽어 온다.

```python
def add(file_path):
    count = 0
    with open(file_path, 'rt', encoding='UTF-8') as file:
        json_data = json.load(file)
        file = open("labeling_data.txt", "a", encoding='UTF-8')
        for i in json_data['SJML']['text']:
            text = i['content'].split('다.')
            text = text[0] + '다.'
            if text.find('?') >= 0:
                text = text.split('?')
                text = text[0] + '?'
            elif text.find('!') >= 0:
                text = text.split('!')
                text = text[0] + '!'
            tx = ''
            lab = text.split(' ')
            for j in lab: #단어 하나씩 들어가 있음
                lab_temp = []
                for k in j: #단어 하나의 글자 하나하나 접근
                    k = 'I ' #해서 I로 바꿔 버림
                    lab_temp.append(k)
                lab_temp[0] = 'B '
                result = ''.join(s for s in lab_temp)
                tx = tx + result
            text = text.replace(' ', '')
            aaa = ''
            for t in text:
                aaa = aaa + t + ' '
            aaa = aaa[0:-1]
            file.write(aaa + '\t ' + tx + '\n')
            count += 1
    file.close()
    return count
```

10개의 파일로도 각각 문장이 1000개가 안 되는 경우가 있어, 추가로 plus(file_path, space) 함수로 메인에서 각각 문장 1000개에서 부족한 양만큼을 매개변수 space로 전달한 후, 함수 내에서 space만큼 파일 더 읽어 온다.

```python
def plus(file_path, space):
    count = 0
    with open(file_path, 'rt', encoding='UTF-8') as file:
        json_data = json.load(file)
        file = open("labeling_data.txt", "a", encoding='UTF-8')
        for i in json_data['SJML']['text']:
            text = i['content'].split('다.')
            text = text[0] + '다.'
            if text.find('?') >= 0:
                text = text.split('?')
                text = text[0] + '?'
            elif text.find('!') >= 0:
                text = text.split('!')
                text = text[0] + '!'
            tx = ''
            lab = text.split(' ')
            for j in lab: #단어 하나씩 들어가 있음
                lab_temp = []
                for k in j: #단어 하나의 글자 하나하나 접근
                    k = 'I ' #해서 I로 바꿔 버림
                    lab_temp.append(k)
                lab_temp[0] = 'B '
                result = ''.join(s for s in lab_temp)
                tx = tx + result
            text = text.replace(' ', '')
            aaa = ''
            for t in text:
                aaa = aaa + t + ' '
            aaa = aaa[0:-1]
            file.write(aaa + '\t ' + tx + '\n')
            count += 1
            if count == 1000 - space:
                break
    file.close()
    return count
```
