```
import numpy as np
from hmmlearn import hmm

states = ["Rainy", "Sunny"]
n_states = len(states)

observations = ["walk", "shop", 'clean']
n_observations = len(observations)

# ì‹œì‘ í™•ë¥ 
start_probability = np.array([0.6, 0.4])

# ì „ì´ í™•ë¥ 
transition_probability = np.array([
  # R    S
  [0.7, 0.3], # R
  [0.4, 0.6]  # S
])

# ê´€ì¸¡ í™•ë¥  observation_probablility
emission_probability = np.array([
  # W    S    C
  [0.1, 0.4, 0.5], # R
  [0.6, 0.3, 0.1]  # S
])

# ëª¨ë¸ ë§Œë“¤ê¸°
# MLE 
model = hmm.CategoricalHMM(n_components=n_states)
model.startprob_ = start_probability
model.transmat_ = transition_probability
model.emissionprob_ = emission_probability

# ê´€ì¸¡ ì…ë ¥
#        W  W  C  S
input = [0, 0, 2, 1]

# HMM ëª¨ë¸ í˜¸ì¶œ
hmm_input = np.atleast_2d(input).T # hmmì— ìµœì í™”ëœ 2D ( 4 * 1 ) í–‰ë ¬ë¡œ ë§Œë“¦ 
logprob, sequence = model.decode(hmm_input)

# ë¹„í„°ë¹„ search -> stateì˜ ì—´ì„ ì°¾ìŒ
print("Input :", ", ".join(map(lambda x: observations[x], input))) # string ê°ì²´ì˜ method 'join' 
print("Output:", ", ".join(map(lambda x: states[x], sequence))) # map: ë¦¬ìŠ¤íŠ¸ì˜ í•˜ë‚˜í•˜ë‚˜ ë„ì§‘ì–´ ë‚´ëŠ” ê²ƒ(ë°˜í™˜) -> í•˜ë‚˜ì”© ë‹¤ join / like forë¬¸
# negative log likelyhoodëŠ” í™•ë¥ ì´ í´ìˆ˜ë¡ ë‚®ì•„ì§ -> loss?
print("Prob. :", logprob)
```
