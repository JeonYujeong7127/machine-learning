𝝿: 시작에서 R, C, S로 가는 확률

---


sequence labeling 문제

```
import numpy as np
from hmmlearn import hmm

states = ["Rainy", "Sunny"]
n_states = len(states)

observations = ["walk", "shop", 'clean']
n_observations = len(observations)

# 시작 확률
start_probability = np.array([0.6, 0.4])

# 전이 확률
transition_probability = np.array([
  # R    S
  [0.7, 0.3], # R
  [0.4, 0.6]  # S
])

# 관측 확률 observation_probablility
emission_probability = np.array([
  # W    S    C
  [0.1, 0.4, 0.5], # R
  [0.6, 0.3, 0.1]  # S
])

# 모델 만들기
# MLE 
model = hmm.CategoricalHMM(n_components=n_states)
model.startprob_ = start_probability
model.transmat_ = transition_probability
model.emissionprob_ = emission_probability

# 관측 입력
#        W  W  C  S
input = [0, 0, 2, 1]

# HMM 모델 호출
hmm_input = np.atleast_2d(input).T # hmm에 최적화된 2D ( 4 * 1 ) 행렬로 만듦 
logprob, sequence = model.decode(hmm_input)

# 비터비 search -> state의 열을 찾음
print("Input :", ", ".join(map(lambda x: observations[x], input))) # string 객체의 method 'join' 
print("Output:", ", ".join(map(lambda x: states[x], sequence))) # map: 리스트의 하나하나 끄집어 내는 것(반환) -> 하나씩 다 join / like for문
# negative log likelyhood는 확률이 클수록 낮아짐 -> loss?
print("Prob. :", logprob)
```
