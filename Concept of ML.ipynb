```
import numpy as np
import torch

# 데이터 읽기 함수
def load_dataset(file):

  data = np.loadtxt(file)
  print("DATA=",data)
  
  x = data[:,0:-1]
  print("X=",x)
  
  y = np.reshape(data[:,-1],(3,1))
  print("Y=",y)
 
  # 텐서로 변경
  # 텐서: 고차원의 array
  x = torch.tensor(x, dtype=torch.float)
  y = torch.tensor(y, dtype=torch.float)

  return (x, y)

# 텐서를 리스트로 바꾸는 함수
def tensor2list(input_tensor):
  return input_tensor.cpu().detach().numpy().tolist()

x, y = load_dataset("/gdrive/My Drive/machine learning/train.txt")

# 가중치 초기화
w = torch.randn(1)*3.0 # w를 -3~3 사이의 값으로 초기화
print("\nInit: w = {0}\n".format(tensor2list(w)))

# 학습률
alpha = 0.1

# 학습데이터 수
m = y.size(0)

# 전체를 다 읽었을 때의 cycle을 epoch라고 함
# data set = 10000인 경우, batch size = 1000으로 잡으면 
# 1 epoch = 10 batch 
for epoch in range(50):
  cost = 0

  for i in range(m):
    # 이곳을 채우세요.
    cost = cost + ((w * x[i]) - y[i]) * x[i]

  # 이곳을 채우세요.
  cost = 1/m * cost

  w = w - alpha * cost
  
  print("Epoch {0:d}: w = {1}".format(epoch+1,tensor2list(w)))
```
